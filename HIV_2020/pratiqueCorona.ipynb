# Pratiquer sur le Coronavirus

Le [coronavirus](https://en.wikipedia.org/wiki/2019%E2%80%9320_Wuhan_coronavirus_outbreak) (ou plus précisément le 2019-nCoV) est le sujet le plus médiatisé ces derniers jours (2020-02-05). Il est à l'origine de l'épidémie de coronavirus de Wuhan en Chine.

<!--  -->

## Question
En vous basant sur [les données](https://github.com/nmeraihi/data/blob/master/2019_nCoV_data.csv) de l'évolution de ce virus datant du 02/03/2020. Ou  les plus à jour que vous pouvez trouver [ici](https://docs.google.com/spreadsheets/d/1wQVypefm946ch4XDp37uZ-wartW4V7ILdg-qYiDXUHM/htmlview?usp=sharing&sle=true#), reproduisez les chiffres, les tableaux ainsi que les graphiques se trouvant dans [tableau de bord suivant](https://gisanddata.maps.arcgis.com/apps/opsdashboard/index.html#/bda7594740fd40299423467b48e9ecf6). 

Vous pouvez reproduire ces chiffres et tableaux avec la librairie `dplyr`. Pour ce qui des graphiques, vous pouvez utiliser `ggplot2`. Pour la carte, la libraire `leaflet` est votre meilleure amie!
