L'une des stratégies de négociation les plus anciennes et les plus simples qui existent est celle qui utilise une moyenne mobile de la série chronologique des prix (ou des rendements) pour représenter la tendance récente du prix.

L'idée est assez simple, mais puissante : si nous utilisons une moyenne mobile de 42 jours (ou 252) de notre série chronologique de prix, alors une partie importante du bruit quotidien des prix aura été "moyennée". Ainsi, nous pouvons observer de plus près le comportement à long terme de l'actif.

Calculons à nouveau les moyennes mobiles *simple moving averages (SMA)* comme suit. Rappelez-vous, encore une fois, que lors du calcul des M jours SMA, les premiers M-1 ne sont pas valables, car M prix sont nécessaires pour le premier point de données des moyennes mobiles.


$$\text{SMA}_{n}=\frac{\text{Prix}_{1}+\text{Prix}_{2}+\ldots+\text{Prix}_{n}}{n}$$

La moyenne mobile peut être calculé sous `Python` avec la fonction [roling](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rolling.html).

```python
import pandas as pd
```

```python
tesla[['Close','42d']].plot(figsize=(8, 5));
```
